```python
# This prompt tells the AI how it should behave.
# I want it to act like a friendly health assistant,
# but never replace a real healthcare professional.

SYSTEM_PROMPT = """
You are a friendly medical assistant.

Follow these rules carefully:

1. Only answer general health-related questions.

2. Explain things using simple language.

3. Never diagnose diseases.

4. Never prescribe medication.

5. Never recommend medicine dosages.

6. If someone describes an emergency,
tell them to seek immediate medical attention.

7. If you're unsure, encourage the user
to speak with a qualified healthcare professional.

Keep your responses friendly, supportive,
and easy to understand.
"""
```

```python
# Some questions shouldn't be answered by an AI.
# I'll check for common emergency or high-risk situations.

unsafe_keywords = [

    "suicide",
    "kill myself",
    "overdose",
    "heart attack",
    "stroke",
    "can't breathe",
    "difficulty breathing",
    "emergency",
    "prescription",
    "dosage",
    "dose",
    "pregnancy",
    "pregnant"

]
```

```python
# This function checks whether the user's question
# contains any keywords that require professional help.

def is_safe(question):

    question = question.lower()

    for keyword in unsafe_keywords:

        if keyword in question:
            return False

    return True
```

```python
# This function sends the user's question to Gemini.
# Before doing that, it checks whether the question is safe.

def ask_health_bot(question):

    # Stop unsafe questions before sending them to the model.
    if not is_safe(question):

        return (
            "I'm sorry, but this question may require professional medical advice.\n\n"
            "Please contact a qualified healthcare provider or your local emergency services if this situation is urgent."
        )

    prompt = f"""
{SYSTEM_PROMPT}

User Question:
{question}
"""

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )

    return response.text
```

```python
print(ask_health_bot("What causes a sore throat?"))
```

```python
print(ask_health_bot("Is paracetamol safe for children?"))
```

```python
print(ask_health_bot("I think I'm having a heart attack."))
```

```python
# This loop lets the user chat with the chatbot until
# they decide to quit.

print("=" * 60)
print("General Health Chatbot")
print("Type 'quit' whenever you want to exit.")
print("=" * 60)

while True:

    user_question = input("\nYou: ")

    if user_question.lower() == "quit":

        print("\nChatbot: Stay safe and take care!")
        break

    reply = ask_health_bot(user_question)

    print("\nChatbot:\n")

    print(textwrap.fill(reply, width=90))
```
